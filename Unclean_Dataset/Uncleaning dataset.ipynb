# Uncleaning The Dataset

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('pokemon.csv')
dataset
```

**Replacing integers by strings**

```python
print(dataset.groupby('against_dragon').groups.keys())
```

```python
dataset['against_dragon'] = dataset['against_dragon'].replace([0.0,0.5,1.0,2.0],['extremly_weak','weak','strong','very_strong'])
dataset
```

```python
print(dataset.groupby('against_ghost').groups.keys())
```

```python
dataset['against_ghost'] = dataset['against_ghost'].replace([0.0,0.5,1.0,2.0,4.0],['extremly_weak','weak','strong','very_strong','powerful'])
dataset
```

```python
print(dataset.groupby('against_steel').groups.keys())
```

```python
dataset['against_steel'] = dataset['against_steel'].replace([0.25,0.5,1.0,2.0,4.0],['very_weak','weak','strong','very_strong','powerful'])
dataset
```

**Adding Duplicates**

```python
df1=dataset.iloc[4]
df2=dataset.iloc[100]
df3=dataset.iloc[205]
df4=dataset.iloc[403]
dataset=dataset.append(df1,ignore_index=True)
dataset=dataset.append(df2,ignore_index=True)
dataset=dataset.append(df3,ignore_index=True)
dataset=dataset.append(df4,ignore_index=True)
dataset
```
